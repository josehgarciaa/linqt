```python
import kwant
import matplotlib.pyplot as plt
import numpy as np
from kwant import wraparound
from matplotlib import pyplot
from numpy import linalg as LA
from numpy import kron
import tinyarray
import time
import itertools
import sys
import os
sys.path.append('/data/jgarcia/codes/linqt/utilities/wannier2sparse/utilities/')


#Custom functions

import wannier_tools  as wt
from kwant2wannier import syst2txt, geo2txt
import graphene

```

```python
syst = graphene.make_bulk_graphene()                 
syst_name = "WS2+graphene+CrGeTe"

hopping_params = {
                    "fermi_energy":0.0,\
                    "staggered_potential":1.147/1000.,\
                    "ferro_exchange" :(0.0,0.0,-4.56/1000.),\
                    "antif_exchange" :(0.0,0.0,-0.01/1000.),
                    "hopping":2.6,\
                    "lambda_rashba":-0.467/1000.,\
                    "lambda_intrinsic":-0.018/1000,\
                    "lambda_valleyZeeman" :1.166/1000.,\
                    "lambda_valleyExchange" :0.158/1000.,
                };
site_dict = {0:"C0_s+_",1:"C1_s+_",2:"C2_s+_",3:"C3_s+_", 4:"C4_s+_", 5:"C5_s+_",
             6:"C0_s-_",7:"C2_s-_",8:"C3_s-_",9:"C4_s-_",10:"C0_s-_",11:"C1_s-_"};
syst2txt(syst,syst_name,hopping_params);
geo2txt(syst, graphene.lattice_vectors,site_dict, syst_name );
```

```python
label = syst_name;
wan_file =label+"_hr.dat"
xyz_file =label+".xyz"
uc_file  =label+".uc"
lat_vec = np.loadtxt(uc_file); 
numOrbs,OrbsID, xyz_coord = wt.load_xyz(xyz_file)
hvec,rows,cols,values = wt.load_wannier_file(wan_file)


num_points = 100;
band_paths = [ ('M1/4\'',num_points,(0.0,-0.001,0)),('G',num_points,(0.0,0.0,0)),('M1/4\'',num_points,(0.0,0.001,0))]

kpoints,bands,label_index = wt._compute_band_structure(numOrbs,hvec,rows,cols,values, band_paths);
#kpoints= np.dot(kpoints,frac2car_rec)
kaxis= np.cumsum( np.insert( np.linalg.norm(kpoints[1:]-kpoints[:-1],axis=1),0,0) )

for band in bands:
    plt.plot(kaxis,band  )

plt.ylim(-0.01,0.01)
plt.show();
```
