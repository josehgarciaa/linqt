```python
import matplotlib.pyplot as plt
import numpy as np
import kwant
import sys
from matplotlib.ticker import MultipleLocator

from numpy import linalg as LA
from kwant import wraparound

#Custom functions
sys.path.append('/data/jgarcia/codes/linqt/utilities/wannier2sparse/utilities/')
import wannier_tools  as wt
from kwant2wannier import syst2txt, geo2txt
import graphene
```

```python
lat_const = 0.25;
vF = 0.7493;
hop = 2.0*graphene.hbar*vF/np.sqrt(3)/lat_const
lattice_vectors = graphene.get_lattice_vectors( lattice_constant = lat_const)
syst_name = "graphene+MoTe2-SC5x7-NOSOC"
hopping_params = {
                    "fermi_energy": 0.,\
                    "staggered_potential": 0*0.0478e-3,\
                    "ferro_exchange" : ( 0.0, 0.0, 0.0 ),\
                    "antif_exchange" : ( 0.0, 0.0, 0.0 ),
                    "hopping": hop,\
                    "lambda_rashba":  0.0,\
                    "lambda_valleyZeeman" : 0.0,\
                    "lambda_valleyExchange" : 0.0,
                };
orb_name = {0:"C0",1:"C1" }; 

syst = graphene.make_bulk_graphene(lat_const=lat_const)                 
syst2txt(syst,syst_name,hopping_params);
geo2txt(syst, lattice_vectors,orb_name, syst_name );
```

```python
#Load the system as a wannier system
ws = wt.wannier_system(syst_name);

num_points = 20;
K0      = np.array([2.0/3.0,1.0/3.0,0]);
K1      = np.array([1.0/3.0,2.0/3.0,0]);
delta_k = 40e-2*np.array([1,1,0])*np.sqrt(3)/4/np.pi/4 /2

#Create a subplot for getting both cones and the spin projections
fig, axs = plt.subplots(2, 2, sharex='all',figsize=(8, 6), dpi=80)
plt.subplots_adjust(bottom=0.1, right=1.3, top=0.9)
#axs[0, 0].plot(x, y)
#axs[0, 0].set_title('Axis [0,0]')
#axs[0, 1].plot(x, y, 'tab:orange')
#axs[0, 1].set_title('Axis [0,1]')
#axs[1, 0].plot(x, -y, 'tab:green')
#axs[1, 0].set_title('Axis [1,0]')
#axs[1, 1].plot(x, -y, 'tab:red')
#axs[1, 1].set_title('Axis [1,1]')
#for ax in axs.flat:
#    ax.set(xlabel='x-label', ylabel='y-label')
# Hide x labels and tick labels for top plots and y ticks for right plots.
#for ax in axs.flat:
#    ax.label_outer()

#Compute Band structure using the wannier hamiltonian
for i,KP in enumerate([K0,K1]):
    band_paths = [ ('-delta_X',1,KP-delta_k),('0',num_points,KP),('delta_X\'',num_points,KP+delta_k)  ];
    ws.compute_band_structure(band_paths, fermi_energy = 0.0, proj_op=ws.spin_operator('z'), ax=axs[0, i] );
    ymin,ymax=-100e-3,100e-3;
    yticks = np.linspace(ymin, ymax ,10);
    axs[0, i].set_ylim(ymin,ymax);
    axs[0, i].set_yticks( yticks );
    axs[0, i].set_yticklabels(list(map(str,1000*yticks)));
  # axs[0, i].set_aspect(aspect=4)
    axs[0, i].grid()

#Compute Band structure using the wannier hamiltonian
for i,KP in enumerate([K0,K1]):
    band_paths = [ ('-delta_X',1,KP-delta_k),('0',num_points,KP),('delta_X\'',num_points,KP+delta_k)  ];
    ws.compute_band_structure(band_paths, fermi_energy = 0.0, proj_op=ws.spin_operator('z'),plot_proj=True, ax=axs[1, i] );
   # yticks = np.linspace(-1.0, 1.0 ,5);
   # axs[1, i].set_ylim(-1.1,1.1);
   # axs[1, i].set_yticks( yticks );
   # axs[1, i].set_yticklabels(list(map(str,yticks)));
   # axs[1, i].set_aspect(aspect=0.01)
    axs[1, i].grid()





#Compute Band structure For continuous model
kpoints = np.dot( ws.get_kpoints(band_paths) - KP , np.transpose(ws.Momentum_Rec2AbsMatrix())  );
kaxis   = ws.bandsXaxis (kpoints, absolute_coords=True );


eigenvalues = list();
for  kp in kpoints:
    Hk = graphene.dirac_model(kp,hopping_params,lat_const=lat_const);
    eigenvalues.append( np.linalg.eigvalsh( Hk ) );
bands = np.transpose(eigenvalues);
for i,band in enumerate(bands):
    axs[0, 0].plot(kaxis,band , color = 'red');


#ymin,ymax=-8e-3,8e-3
#yticks = np.linspace(ymin, ymax ,9);
#ax.set_ylim(ymin,ymax);
#ax.set_yticks( yticks );
#ax.set_yticklabels(list(map(str,1000*yticks)));
#ax.set_aspect(3.0)
```
